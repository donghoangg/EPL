```python
import pandas as pd

# Đường dẫn đến các file
squad_info_path = r"E:\project 1\data_clean\squad_info.xlsx"
season_path = r"E:\project 1\data_clean\season.xlsx"
club_path = r"E:\project 1\data_clean\club.xlsx"

try:
    # 1. Đọc các file Excel vào các DataFrame
    df_squad = pd.read_excel(squad_info_path)
    df_season = pd.read_excel(season_path)
    df_club = pd.read_excel(club_path)

    # 2. Lấy Season_ID bằng cách merge với df_season
    # 'how="left"' để giữ lại tất cả các dòng từ df_squad
    df_merged = pd.merge(df_squad, df_season, on="Season", how="left")

    # 3. Lấy Club ID mới bằng cách merge với df_club
    # Sử dụng left_on và right_on vì tên cột không giống nhau
    # left_on='Club ID' (từ file squad_info) và right_on='Squad' (từ file club)
    df_merged = pd.merge(df_merged, df_club, left_on="Club ID", right_on="Squad", how="left")

    # 4. Chọn và sắp xếp lại các cột cuối cùng
    # Lấy 'Club ID_y' là cột Club ID mới từ file club.xlsx
    # và đổi tên nó thành 'Club ID'
    final_df = df_merged[['Season_ID', 'Club ID_y', 'Num player', 'Average_Age']]
    final_df = final_df.rename(columns={"Club ID_y": "Club_ID"})

    # 5. Lưu DataFrame kết quả trở lại file squad_info.xlsx
    # index=False để không ghi chỉ số của DataFrame
    final_df.to_excel(squad_info_path, index=False)

    print(f"Tệp {squad_info_path} đã được cập nhật thành công.")
    print("Các cột đã được thay thế và sắp xếp lại theo yêu cầu.")

except FileNotFoundError as e:
    print(f"Lỗi: Không tìm thấy tệp. Vui lòng kiểm tra lại đường dẫn: {e.filename}")
except KeyError as e:
    print(f"Lỗi: Không tìm thấy cột trong tệp Excel. Vui lòng kiểm tra tên cột: {e}")
except Exception as e:
    print(f"Đã xảy ra lỗi không mong muốn: {e}")
```

```python
import pandas as pd

# Đường dẫn đến file Excel của bạn
file_path = r"E:\project 1\data_clean\squad_info.xlsx"


try:
    # 1. Đọc file Excel vào DataFrame
    df = pd.read_excel(file_path)

    # 2. Tạo cột 'Club_Season_ID' mới
    # Đảm bảo cả hai cột đều là kiểu chuỗi (string) trước khi gộp
    # để tránh lỗi nếu có cột nào đó đang ở dạng số.
    df['Club_Season_ID'] = df['Club_ID'].astype(str) + '-' + df['Season_ID'].astype(str)

    # 3. (Tùy chọn) Sắp xếp lại thứ tự các cột để đặt cột mới ở đầu
    # Lấy danh sách tất cả các cột
    cols = df.columns.tolist()
    # Di chuyển cột cuối cùng (là 'Club_Season_ID') lên vị trí đầu tiên
    cols = [cols[-1]] + cols[:-1]
    # Áp dụng thứ tự cột mới cho DataFrame
    df = df[cols]

    # 4. Lưu DataFrame đã cập nhật trở lại file Excel
    # index=False để không ghi thêm cột chỉ số không cần thiết
    df.to_excel(file_path, index=False)

    print(f"Đã thêm thành công cột 'Club_Season_ID' vào file: {file_path}")
    print("File đã được lưu lại với cột mới ở vị trí đầu tiên.")

except FileNotFoundError:
    print(f"Lỗi: Không tìm thấy tệp tại đường dẫn: {file_path}")
except KeyError as e:
    print(f"Lỗi: Không tìm thấy cột cần thiết ({e}) trong tệp. Vui lòng kiểm tra lại tên cột.")
except Exception as e:
    print(f"Đã xảy ra một lỗi không mong muốn: {e}")
```
